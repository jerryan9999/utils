```python
import pymongo
from pprint import pprint
```

```python
client = pymongo.MongoClient()
collection = client['Zillow']['Rooms']
```

```python
pprint(collection.find_one({},{'price_history_accumulate':0,'description':0,'additional_features':0}))
```

```python
collection_ne = client['Zillow']['neighborhoods']
neighborhood = collection_ne.find_one( { 'geometry': { '$geoIntersects': { '$geometry': { 'type': "Point", 'coordinates': [ -118.267551, 34.078166 ] } } } } )
```

```python
pprint(neighborhood)
```

```python
pprint(collection.find({'loc':{ '$geoWithin': { '$geometry': neighborhood['geometry'] } } }).count())
```
